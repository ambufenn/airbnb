Data cleansing is an important step in the data analysis process. Here are some general steps for cleansing data:

    Identify and remove duplicates: Check for any duplicate rows or observations in the dataset and remove them.

    Identify and handle missing values: Check for any missing values in the dataset and decide on the appropriate way to handle them. Options include deleting the rows with missing values, imputing missing values, or using techniques such as mean or median imputation.

    Identify and handle outliers: Check for any outliers in the dataset and decide on the appropriate way to handle them. Options include removing them, transforming the data, or using statistical methods to detect and adjust for outliers.

    Standardize and normalize data: Convert data into a common scale or range to allow for comparison between variables. This can involve standardizing or normalizing the data.

    Handle categorical data: Convert categorical data into a numerical format that can be used in analyses. This can involve techniques such as one-hot encoding.

    Check data integrity: Check for any inconsistencies or errors in the dataset that may affect analysis. This can involve cross-checking data against external sources or performing data validation checks.

    Verify data quality: Check that the data meets quality standards, such as accuracy, completeness, and consistency.

These are general steps for cleansing data, and the specific steps and techniques used may vary depending on the specific dataset and analysis.

cara bersihkan outlier

```python


# assuming df3 is a pandas DataFrame with a 'price' column
q1 = df3['price'].quantile(0.25)
q3 = df3['price'].quantile(0.75)

# calculate upper and lower limits for outliers detection
iqr = q3 - q1
upper_limit = q3 + 1.5 * iqr
lower_limit = q1 - 1.5 * iqr
```

```python
#filter price outlier

df3 = df3[df3['price'] > lower_limit]
df3 = df3[df3['price'] < upper_limit]

```

# 1.  BagaimanaÂ distribusiÂ hargaÂ listingÂ airbnbÂ diÂ Singapura?

```python

# menentukan rentang harga
bins = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]

# menghitung jumlah listing pada setiap rentang harga
hist = pd.cut(df3['price'], bins=bins).value_counts(sort=False)

# membuat diagram distribusi harga
plt.bar(hist.index.astype(str), hist.values)
plt.xticks(rotation=45)
plt.xlabel('Rentang Harga (SGD)')
plt.ylabel('Jumlah Listing')
plt.title('Distribusi Harga Listing Airbnb di Singapura')
plt.show()
```

# BagaimanaÂ trendÂ hargaÂ padaÂ NeighborhoodÂ ?

```python
df
```

```python
# Menentukan variabel Neighborhood dan price
var = ["neighbourhood", "price"]

# Mengelompokkan data berdasarkan Neighborhood dan menghitung rata-rata harga
grouped_df = df[var].groupby("neighbourhood").mean()

# Mengurutkan data secara menurun berdasarkan harga
sorted_df = grouped_df.sort_values(by="price", ascending=False)

# Membuat grafik bar chart
plt.figure(figsize=(15, 7))
plt.bar(sorted_df.index, sorted_df["price"])
plt.xticks(rotation=90)
plt.title("Trend Harga pada Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Harga")
plt.show()

```

BagaimanaÂ trendÂ penyewaanÂ dariÂ 01-01-2018Â sampaiÂ denganÂ 22-09-2022Â ?Â Intermediate


```python
# Agregasi jumlah penyewaan per bulan
df_monthly = df.groupby(pd.Grouper(key='date', freq='M'))['id'].count().reset_index()

# Buat grafik
plt.figure(figsize=(12,6))
sns.lineplot(x="date", y="id", data=df_monthly)
plt.title("Trend Penyewaan 01-01-2018 sampai dengan 22-09-2022")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Penyewaan")
plt.show()
```

Berdasarkan grafik yang dihasilkan, terlihat bahwa tren penyewaan akomodasi di Singapura mengalami kenaikan yang signifikan pada tahun 2019 dan awal tahun 2020, tetapi kemudian mengalami penurunan yang tajam pada pertengahan tahun 2020, yang diduga terkait dengan pandemi COVID-19 dan pembatasan perjalanan yang dihasilkan dari pandemi tersebut.

Setelah penurunan tajam, terjadi sedikit peningkatan pada awal tahun 2021, tetapi kemudian turun kembali pada pertengahan tahun 2021. Namun, pada akhir tahun 2021 hingga awal 2022, terjadi kenaikan tajam dalam tren penyewaan, yang kemungkinan terkait dengan pelonggaran pembatasan perjalanan dan peningkatan vaksinasi.

Secara keseluruhan, grafik ini menunjukkan bahwa trend penyewaan akomodasi di Singapura sangat dipengaruhi oleh faktor eksternal seperti pembatasan perjalanan dan pandemi COVID-19, serta faktor-faktor lain seperti musim liburan dan acara besar.

host_idÂ danÂ host_nameÂ apaÂ yangÂ mempunyaiÂ jumlahÂ listingÂ terbanyakÂ danÂ palingÂ sedikitÂ 

```python
# Mendapatkan host_id dan host_name yang mempunyai jumlah listing terbanyak
host_id_terbanyak = df["host_id"].value_counts().idxmax()
host_name_terbanyak = df.loc[df["host_id"] == host_id_terbanyak, "host_name"].iloc[0]

# Mendapatkan host_id dan host_name yang mempunyai jumlah listing paling sedikit
host_id_paling_sedikit = df["host_id"].value_counts().idxmin()
host_name_paling_sedikit = df.loc[df["host_id"] == host_id_paling_sedikit, "host_name"].iloc[0]

# Menampilkan hasil
print("Host_id dengan jumlah listing terbanyak: ", host_id_terbanyak)
print("Host_name dengan jumlah listing terbanyak: ", host_name_terbanyak)
print("Host_id dengan jumlah listing paling sedikit: ", host_id_paling_sedikit)
print("Host_name dengan jumlah listing paling sedikit: ", host_name_paling_sedikit)
```
